## Advanced Encryption Standard(AES)

- It was Developed by  National Institute of Standards and Technology (NIST) in 2001
- AES operates on 128-bit blocks = 16 bytes, arranged as a 4×4 State matrix.
- here matrix is called state
- The algoritm supports differnt key sizes
- Rounds based on the key sizes

| Key Size (bits) | Number of Rounds |
|-----------------|------------------|
| 128             | 10               |
| 192             | 12               |
| 256             | 14               |



```
Algorithm :
Step 1: Start.

Step 2: Read the 128-bit plaintext block and the secret key (128/192/256 bits).

Step 3: Perform Key Expansion on the original key to generate all Round Keys.

Step 4: Arrange the 16-byte plaintext into a 4×4 State matrix.

Step 5: Perform the Initial Round:
        • AddRoundKey(State, RoundKey[0])

Step 6: For each of the next (Nr – 1) rounds, repeat:
        a) SubBytes: Substitute every byte using the AES S-box.
        b) ShiftRows: Circular left-shift each row by its row index.
        c) MixColumns: Transform each column using fixed GF(2⁸) matrix.
        d) AddRoundKey: XOR the State with the current round key.

Step 7: Perform the Final Round (without MixColumns):
        a) SubBytes
        b) ShiftRows
        c) AddRoundKey(State, RoundKey[Nr])

Step 8: Convert the final State matrix into a 128-bit ciphertext block.

Step 9: Display/return the ciphertext.

Step 10: Stop.
```
code
```python
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.primitives import padding
from cryptography.hazmat.backends import default_backend
import os
import binascii

def bytes_to_hex(byte_data: bytes) -> str:
    """Convert bytes to lowercase hex string"""
    return binascii.hexlify(byte_data).decode("utf-8")

def main():
    # Plaintext message
    message = "AES still rocks!!"

    # Generate a random 128-bit key (16 bytes)
    key = os.urandom(16)

    # Generate a random 128-bit IV for CBC mode
    iv = os.urandom(16)

    # Convert text to bytes
    plaintext = message.encode("utf-8")

    # PKCS7 padding
    padder = padding.PKCS7(algorithms.AES.block_size).padder()
    padded_plaintext = padder.update(plaintext) + padder.finalize()

    # Encrypt using AES-CBC
    encryptor = Cipher(
        algorithms.AES(key),
        modes.CBC(iv),
        backend=default_backend()
    ).encryptor()

    ciphertext = encryptor.update(padded_plaintext) + encryptor.finalize()

    print("Key (hex):", bytes_to_hex(key))
    print("IV (hex):", bytes_to_hex(iv))
    print("Plaintext:", message)
    print("Ciphertext (hex):", bytes_to_hex(ciphertext))

    # Decrypt
    decryptor = Cipher(
        algorithms.AES(key),
        modes.CBC(iv),
        backend=default_backend()
    ).decryptor()

    decrypted_padded = decryptor.update(ciphertext) + decryptor.finalize()

    # Remove padding
    unpadder = padding.PKCS7(algorithms.AES.block_size).unpadder()
    decrypted = unpadder.update(decrypted_padded) + unpadder.finalize()

    print("\nDecrypted:", decrypted.decode("utf-8"))
    print("Decrypted (hex):", bytes_to_hex(decrypted))

if __name__ == "__main__":
    try:
        main()
    except Exception as e:
        print("Error:", e)
        
```
